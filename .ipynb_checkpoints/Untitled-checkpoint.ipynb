```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nba_api.stats.endpoints import leaguedashplayerstats  # Remove advanced stats if not available

# Fetch 2023-24 season player stats
def get_nba_data():
    stats = leaguedashplayerstats.LeagueDashPlayerStats(season='2023-24', per_mode_detailed='PerGame')
    df = stats.get_data_frames()[0]
    return df

# Load data
df = get_nba_data()  # Corrected function name

# Select key columns
columns = ["PLAYER_NAME", "TEAM_ABBREVIATION", "GP", "PTS", "AST", "REB", "STL", "BLK", "TOV", "FG_PCT", "FT_PCT"]
df = df[columns]

# Convert percentages to readable format
df["FG_PCT"] *= 100
df["FT_PCT"] *= 100

# Sort by top scorers
top_scorers = df.sort_values(by="PTS", ascending=False).head(10)

# Plot Top 10 Scorers with hue for color mapping
plt.figure(figsize=(10, 5))
sns.barplot(y=top_scorers["PLAYER_NAME"], x=top_scorers["PTS"], hue=top_scorers["PLAYER_NAME"], palette='coolwarm', dodge=False)
plt.xlabel("Points Per Game")
plt.ylabel("Player")
plt.title("Top 10 Scorers of the 2023-24 Season")
plt.legend(title="Player", loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()

# Save cleaned data
print(df.columns)  # This will display the columns after cleaning the data
```

## Summary of Analysis:
The current analysis focuses on advanced statistics from the 2023-2024 NBA season, particularly evaluating player performance through key metrics. The following steps have been completed:

Data Extraction: The nba_api was used to extract player stats for the 2023-24 season, including metrics like points per game (PTS), assists (AST), rebounds (REB), steals (STL), blocks (BLK), turnovers (TOV), and shooting percentages (FG_PCT, FT_PCT).
Data Preparation: The data was cleaned and transformed by:
Selecting relevant columns for analysis.
Converting shooting percentages (FG% and FT%) into more readable formats by multiplying by 100.
Visualization: A barplot was created to visualize the Top 10 Scorers of the 2023-24 season, showing points per game. This plot helps to easily identify the top players in terms of scoring.
Key Insights:
From this analysis, you should be able to quickly see which players are leading the scoring race. The visualization gives a clear, intuitive representation of how different players perform in comparison to each other in terms of points per game.
